# Churn cleaning workflow in Python

```python
# Import Libraries
import pandas as pd
import numpy as np

df = pd.read_excel("Telco_customer_churn.xlsx")

df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
# Drop Unnecessary Columns
df.drop(columns=[
    "Count",
    "Country",
    "Lat Long",
    "Latitude",
    "Longitude"
], inplace=True)
```

```python
df["Churn Value"].value_counts()
```

```python
#Tenure Months

def tenure_group(x):
    if x <= 6:
        return "0-6 Months"
    elif x <= 12:
        return "6-12 Months"
    elif x <= 24:
        return "1-2 Years"
    else:
        return "2+ Years"

df["Tenure_Group"] = df["Tenure Months"].apply(tenure_group)
```

```python
df["Tenure_Group"].value_counts()
```

```python
#Revenue Segmentation
df["Revenue_Segment"] = pd.qcut(
    df["Monthly Charges"],
    q=3,
    labels=["Low", "Medium", "High"]
)
df["Revenue_Segment"].value_counts()
```

```python
df["Churn Reason"].fillna("Not Churned", inplace=True)
```

```python
# Quick Business Insights

#overall Churn Rate
df["Churn Value"].mean() * 100


```

```python

#churn by contract type
df.groupby("Contract")["Churn Value"].mean() * 100
```

```python
#churn by tenure group
df.groupby("Tenure_Group")["Churn Value"].mean() * 100
```

```python
df.to_csv("churn_cleaned_final.csv", index=False)
```
